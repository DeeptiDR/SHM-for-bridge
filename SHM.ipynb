### 2) Now we will do our analysis on each individual files.
#### For MGCplus_1.csv, Fast Fourier Transform is performed on the time series data. The time domain curve and the frequency domain curve is shown as follows:

```python
import numpy as np
from numpy.fft import fft, fftfreq, ifft
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('MGCplus_1.csv')

n = len(df.Time) #length of my dataset or sampled signal of length n
T = 0.005
Lx = n*T

x = np.linspace(0, Lx, n)
y=df.MGCplus_1

freqs=fftfreq(n,T) #Return the Discrete Fourier Transform sample frequencies.
mask=freqs>0
fft_vals=fft(y)
fft_theo=2.0*np.abs(fft_vals/n) #abs give only vector of complex term #1/n is a normalization factor. 
#fft_theo calculates the magnitude of signal at corresponding frequencies.

plt.figure(1)
plt.title('Original Signal')
plt.plot(x, y, color='xkcd:salmon', label='original')
plt.xlabel('Time (sec)', fontsize=13)
plt.ylabel('acceleration (m/s2)', fontsize=13)
plt.legend()
plt.grid()

plt.figure(2)
plt.title('True FFT values')
plt.plot(freqs[mask], fft_theo[mask], label='true fft values')
plt.xlabel('Frequency (Hz)', fontsize=13)
plt.ylabel('magnitude', fontsize=13)
plt.legend()
plt.grid()

plt.show()
```

#### Now, from the time series graph (orange), we can notice the sudden peaks at the extreme ends. These are outliers for us.


### 3) Out of many outlier removal techniques, we will use LOF (Local Outlier Factor) algorithm that is available in Scikit learn library.

#### We will use LOF on our entire dataset, it will return -1 for outlier values and 1 for non-outlier values

```python
from sklearn.neighbors import LocalOutlierFactor
df1=pd.read_csv('BR85_C_II.csv', delimiter='\t')
model=LocalOutlierFactor(novelty=True)
model.fit(df1)
print(pd.Series(model.predict(df1)).value_counts())
```

#### There are 37792 non-outlier values and 3908 outlier values in our dataset for each column

### Now for our first accelerometer reading, we will see the corresponding outlier and non-outlier readings as follows

```python
df1=pd.read_csv('BR85_C_II.csv', delimiter='\t')
df=pd.Series(model.predict(df1))
elements=np.array(df)
elements=elements.reshape(elements.shape[0],1)
df=pd.DataFrame({'result':elements[:,0]})
df = pd.concat([df1.Time, df1.MGCplus_1, df.result], axis=1)
df.to_csv('MGCplus_1.csv', index=False)

df=pd.read_csv("MGCplus_1.csv")
df
```

### 4) 'result' column shows '1' for the non-outlier values and '-1' for outlier values.
#### Now outliers are removed entirely as follows

```python
dataF=pd.read_csv('BR85_C_II.csv', delimiter='\t')
df=pd.read_csv('MGCplus_1.csv')
df.loc[df.result == -1, 'MGCplus_1'] = np.NaN
df1=pd.DataFrame({'ModifiedMGCplus_1':df.MGCplus_1})
df=pd.concat([df.Time, dataF.MGCplus_1, df.result, df1.ModifiedMGCplus_1], axis=1)
df.to_csv('MGCplus_1.csv', index=False)

df=pd.read_csv("MGCplus_1.csv")
df
```

#### 'ModifiedMGCplus_1' column contains the values with no outliers

### 5) Linear interpolation method is used to fill the NaN values in 'ModifiedMGCplus_1' column

```python
df=pd.read_csv('MGCplus_1.csv')
df=df['ModifiedMGCplus_1'].interpolate(limit_direction='forward')
df=pd.DataFrame({'interpolated':df})
df.to_csv('interpolated.csv')

df2=pd.read_csv('interpolated.csv')
df=df2['interpolated'].interpolate(limit_direction='backward')
df=pd.DataFrame({'interpolated':df})
df.to_csv('interpolated.csv')


df1=pd.read_csv('MGCplus_1.csv')
df2=pd.read_csv('interpolated.csv')
df=pd.concat([df1.Time, df2.interpolated, df1.MGCplus_1, df1.result, df1.ModifiedMGCplus_1],axis=1)
df.to_csv('MGCplus_1.csv', index=False)

df=pd.read_csv("MGCplus_1.csv")
df
```

#### The column 'Interpolated' contains the interpolated value

### 6) Finally data normalization is performed on interpolated values. The final values are stored in column 'normalized'
#### Normalization= value-mean

```python
df1 = pd.read_csv('MGCplus_1.csv')
std=np.std(df1.interpolated)
mean=np.mean(df1.interpolated)

df=(df1.interpolated-mean)
df2=pd.DataFrame({'normalized':df})
df2.to_csv('normalized.csv')

df1=pd.read_csv('MGCplus_1.csv')
df2=pd.read_csv('normalized.csv')

df=pd.concat([df1.Time, df2.normalized, df1.interpolated, df1.MGCplus_1, df1.result, df1.ModifiedMGCplus_1],axis=1)
df.to_csv('MGCplus_1.csv', index=False)

df=pd.read_csv("MGCplus_1.csv")
df
```

#### 'normalized' column contains the final values after performing -> Outlier removal, Interpolation and Data normalization

### 7) For 'normalized' column of MGCplus_1.csv, Fast Fourier Transform is performed on the time series data. The time domain curve and the frequency domain curve is shown as follows:

```python
df = pd.read_csv('MGCplus_1.csv')

n = len(df.Time) #length of my dataset or sampled signal of length n
T = 0.005
Lx = n*T

x = np.linspace(0, Lx, n)
y=df.normalized

freqs=fftfreq(n,T) #Return the Discrete Fourier Transform sample frequencies.
mask=freqs>0
fft_vals=fft(y)
fft_theo=2.0*np.abs(fft_vals/n) #abs give only vector of complex term #1/n is a normalization factor. 
#fft_theo calculates the magnitude of signal at corresponding frequencies.

plt.figure(1)
plt.title('Original Signal')
plt.plot(x, y, color='xkcd:salmon', label='original')
plt.xlabel('Time (sec)', fontsize=13)
plt.ylabel('acceleration (m/s2)', fontsize=13)
plt.legend()
plt.grid()

plt.figure(2)
plt.title('True FFT values')
plt.plot(freqs[mask], fft_theo[mask], label='true fft values')
plt.xlabel('Frequency (Hz)', fontsize=13)
plt.ylabel('magnitude', fontsize=13)
plt.legend()
plt.grid()

plt.show()
```
